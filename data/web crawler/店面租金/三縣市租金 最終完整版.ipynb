```python
# 設定啟動瀏覽器工具的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")         #最大化視窗
options.add_argument("--disable-popup-blocking ") #禁用彈出攔截

browser = webdriver.Chrome(options = options)

# input 物件網址 撈取網頁資料 return 欄位 
def getData(url):
    
    request_url='https:'+str(url).strip()
    res=requests.get(request_url)
    
    if res.status_code == 200:
        bs=BeautifulSoup(res.text,'html.parser')
        #先宣告變數為NULL 若無撈到資料則寫入NULL
        num='NULL'
        addr='NULL'
        price='NULL'
        size='NULL'
        floor='NULL'
        room_type='NULL'
        history_date='NULL'

        # 利用 beautfiulsoup 的 find function 利用 css selector 定位 並撈出指定資料
        try:
            num=bs.find('i',{'style':'font-style: inherit; color: #999'}).text
        except:
            num='None'
        try:
            addr=bs.find('span',{'class':'addr'}).text
        except:
            addr='None'
        try:
            price=bs.find('div',{'class':'price'}).text.strip().split(' ')[0]
        except:
            price='None'
        try:
            room_attrs=bs.find('ul',{'class':'attr'}).findAll('li')
            for attr in room_attrs:
                if attr.text.split('\xa0:\xa0\xa0')[0]=='坪數':
                    size=attr.text.split('\xa0:\xa0\xa0')[1]
                elif attr.text.split('\xa0:\xa0\xa0')[0]=='樓層':
                    floor=attr.text.split('\xa0:\xa0\xa0')[1]
                elif attr.text.split('\xa0:\xa0\xa0')[0]=='型態':
                    room_type=attr.text.split('\xa0:\xa0\xa0')[1]
        except:
            size='None'
            floor='None'
            room_type='None'
        try:
            htd=bs.find('span',{'class':'ft-rt'}).text
            history_date=str(htd[4:])
        except:
            history_date='None'
        return num,addr,price,size,floor,room_type,history_date
    else:
        return 404,404,404,404,404,404,404
    
def main(outputfile):
    #利用chrome模擬器開啟
    browser.get("https://business.591.com.tw/?type=1&kind=5")
    time.sleep(random.randint(3,7))
    browser.find_element_by_class_name('close').click()
    time.sleep(random.randint(3,7))
    browser.find_element_by_id('areaSelect').click()
    time.sleep(random.randint(3,7))
    browser.find_element_by_xpath("//*[@id='areaBoxNew']/dl[1]/dd/a[1]").click()
    
    #輸入 ESC 關閉google 提示，否則無法點選
    time.sleep(random.randint(3,7))
    browser.find_element_by_class_name('pageNext').send_keys(Keys.ESCAPE) #ECS鍵

    with open(outputfile, 'w', newline='',encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        
        bs = BeautifulSoup(browser.page_source, 'html.parser')
        totalpages = int(bs.find('span', {'class':'TotalRecord'}).text.split(' ')[-2])//30 + 1
        print('Total pages: ', totalpages)

        for i in range(totalpages):
            room_url_list=[] #存放網址list
            bs = BeautifulSoup(browser.page_source, 'html.parser')
            titles=bs.findAll('h3') # h3 放置物件的區塊
            for title in titles:
                room_url=title.find('a').get('href') # 每個物件的 url
                room_url_list.append(room_url)
            time.sleep(random.randint(3,7))

            # ------------- write into csv ------------- #
            for url in room_url_list:
                num,addr,price,size,floor,room_type,history_date = getData(url)
                writer.writerow([num,addr,price,size,floor,room_type,history_date])
            # ------------------------------------------ #
            print('%.2f'%((i+1)/totalpages*100), '%',end='\r') # print out 完成 %數

            #若偵測到無法點選最後一頁則跳出
            if bs.find('a',{'class':'last'}):
                pass
            else:
                #撈取完資料後點選下一頁，並等待 3 秒載入新頁面
                browser.find_element_by_class_name('pageNext').send_keys(Keys.ESCAPE)
                browser.find_element_by_class_name('pageNext').click()
                time.sleep(random.randint(3,7))

try:
    if __name__ == '__main__':
        # -------- configurable parameter -------- #
        output_file_name = '臺北-房仲.csv'
        # ---------------------------------------- #
        main(output_file_name)
except BaseException as e:
    print("錯誤!!")
    print(e)
else:
    print("恭喜!抓完 臺北-房仲 資料!")
finally:
     browser.quit()
```

```python
# 設定啟動瀏覽器工具的選項
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")         #最大化視窗
options.add_argument("--disable-popup-blocking ") #禁用彈出攔截

browser = webdriver.Chrome(options = options)

# input 物件網址 撈取網頁資料 return 欄位 
def getData(url):
    
    request_url='https:'+str(url).strip()
    res=requests.get(request_url)
    
    if res.status_code == 200:
        bs=BeautifulSoup(res.text,'html.parser')
        #先宣告變數為NULL 若無撈到資料則寫入NULL
        num='NULL'
        addr='NULL'
        price='NULL'
        size='NULL'
        floor='NULL'
        room_type='NULL'
        history_date='NULL'

        # 利用 beautfiulsoup 的 find function 利用 css selector 定位 並撈出指定資料
        try:
            num=bs.find('i',{'style':'font-style: inherit; color: #999'}).text
        except:
            num='None'
        try:
            addr=bs.find('span',{'class':'addr'}).text
        except:
            addr='None'
        try:
            price=bs.find('div',{'class':'price'}).text.strip().split(' ')[0]
        except:
            price='None'
        try:
            room_attrs=bs.find('ul',{'class':'attr'}).findAll('li')
            for attr in room_attrs:
                if attr.text.split('\xa0:\xa0\xa0')[0]=='坪數':
                    size=attr.text.split('\xa0:\xa0\xa0')[1]
                elif attr.text.split('\xa0:\xa0\xa0')[0]=='樓層':
                    floor=attr.text.split('\xa0:\xa0\xa0')[1]
                elif attr.text.split('\xa0:\xa0\xa0')[0]=='型態':
                    room_type=attr.text.split('\xa0:\xa0\xa0')[1]
        except:
            size='None'
            floor='None'
            room_type='None'
        try:
            htd=bs.find('span',{'class':'ft-rt'}).text
            history_date=str(htd[4:])
        except:
            history_date='None'
        return num,addr,price,size,floor,room_type,history_date 
    else:
        return 404,404,404,404,404,404,404
    
def main(outputfile):
     #利用chrome模擬器開啟
    browser.get("https://business.591.com.tw/?type=1&kind=5")
    time.sleep(random.randint(3,7))
    browser.find_element_by_class_name('close').click()
    #關閉選取地區pop-up 否則無法點選下一頁
    browser.find_element_by_id('areaSelect').click()
    time.sleep(random.randint(3,7))
    browser.find_element_by_xpath("//*[@id='areaBoxNew']/dl[1]/dd/a[3]").click()
    time.sleep(random.randint(3,7))
    browser.find_element_by_xpath("//*[@id='container']/section[5]/div/div[1]/ul/li[3]/a").click()
    
    #輸入 ESC 關閉google 提示，否則無法點選
    time.sleep(random.randint(3,7))
    browser.find_element_by_class_name('pageNext').send_keys(Keys.ESCAPE) #ECS鍵

    with open(outputfile, 'w', newline='',encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile)
        
        bs = BeautifulSoup(browser.page_source, 'html.parser')
        totalpages = int(bs.find('span', {'class':'TotalRecord'}).text.split(' ')[-2])//50 + 1
        print('Total pages: ', totalpages)

        for i in range(totalpages):
            room_url_list=[] #存放網址list
            bs = BeautifulSoup(browser.page_source, 'html.parser')
            titles=bs.findAll("div", {"class":"cplAddress"}) # div 放置物件的區塊
            for title in titles:
                room_url=title.find('a').get('href') # 每個物件的 url
                room_url_list.append(room_url)
            time.sleep(random.randint(3,7))

            # ------------- write into csv ------------- #
            for url in room_url_list:
                num,addr,price,size,floor,room_type,history_date = getData(url)
                writer.writerow([num,addr,price,size,floor,room_type,history_date])
            # ------------------------------------------ #
            print('%.2f'%((i+1)/totalpages*100), '%',end='\r') # print out 完成 %數

            #若偵測到無法點選最後一頁則跳出
            if bs.find('a',{'class':'last'}):
                pass
            else:
                #撈取完資料後點選下一頁，並等待 3~6 秒載入新頁面
                browser.find_element_by_class_name('pageNext').send_keys(Keys.ESCAPE)
                browser.find_element_by_class_name('pageNext').click()
                time.sleep(random.randint(3,7))

try:
    if __name__ == '__main__':
        # -------- configurable parameter -------- #
        output_file_name = '桃園-歷史資料.csv'
        # ---------------------------------------- #
        main(output_file_name)
except BaseException as e:
    print("錯誤!!")
    print(e)
else:
    print("恭喜!抓完 桃園-歷史 資料!")
finally:
     browser.quit()
```
