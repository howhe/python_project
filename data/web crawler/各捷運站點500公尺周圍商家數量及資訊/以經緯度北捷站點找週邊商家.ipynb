```python
#再讀一次csv把經緯度包成字串
df=pd.read_csv('all_stations_finished.csv',dtype={'lat':str,'lng':str})
df_loc=df['lat']+','+df['lng']
MRT=df_loc.values.tolist()
```

```python
# 抓取各站點500m周圍商家數量及資訊，並放進ids_cafe之建立好的四項欄位(商家:咖啡店)
ids_cafe = {"車站種類":[],"車站編號":[],"車站名稱":[],"CAFE_NUM(500)":[]}
ids_cafe_info=[]
for station in MRT:
    results = []
    query_result = gmaps.places_nearby(keyword='cafe',location=station,radius=500)
    results.extend(query_result['results'])
    while query_result.get('next_page_token'):
        time.sleep(2)
        query_result = gmaps.places_nearby(page_token=query_result['next_page_token'])
        results.extend(query_result['results'])        
    print("找到以北北基桃各站點"+station+"為中心半徑500公尺的咖啡店家數量(google mapi api上限提供60間): "+str(len(results)))
    ids_cafe["CAFE_NUM(500)"].append(str(len(results)))
    for place in results:
        ids_cafe_info.append(place['place_id'])
for sn in station_name:
    ids_cafe["車站名稱"].append(sn)
for st in station_type:
    ids_cafe["車站種類"].append(st)
for sc in station_code:
    ids_cafe['車站編號'].append(sc) 
```
